# 비선형 활성화 함수(Activation function)

비선형 활성화 함수(Activation function): 입력을 받아 수학적 변환을 수행하고 출력을 생성하는 함수

활성화 함수의 종류: 시그모이드, 소프트맥스 등등..

## 활성화 함수의 특징 - 비선형 함수(Nonlinear function)

활성화 함수를 비선형 함수로 사용하는 이유: 선형 함수는 학습 가능한 가중치가 새로 생긴다는 점에서 분명히 의미가 있지만 선형 함수로는 은닉층을 여러번 추가하더라도 1회 추가한 것과 같기 때문에 비선형 함수로 은닉층을 쌓아야 

## 시그모이드 함수(Sigmoid function)와 기울기 소실

인공 신경망의 학습 과정: 순전파(forward propagation) -(예측값과 실제값의 오차)-> 손실 함수(loss function) -(loss를 미분해서 구한 gradient)-> 역전파(back propagation)

시그모이드 함수의 문제점: 미분을 해서 기울기(gradient)를 구할 때 즉 역전파 과정에서 0에 가까운 아주 작은 기울기가 곱해지게 되면, 앞단에는 기울기가 잘 전달되지 않게 됩니다. 다시 말해 매개변수 W가 업데이트 되지 않아 학습이 되지를 않습니다. 결론적으로 시그모이드 함수를 은닉층에서 사용하는 것은 지양됩니다.

## 소프트맥스 함수(Softamx function)

시그모이드 함수나 소프트맥스 함수가 은닉층에서는 사용되지 않지만  
분류 문제의 출력층에 사용됩니다.

## 하이퍼볼릭탄젠트 함수(Hyperbolic tangent function)

시그모이드 함수보다는 기울기 소실 증상이 적은 편입니다. 그래서 은닉층에서 시그모이드 함수보다는 많이 사용됩니다.

## 렐루 함수(ReLU)

f(x) = max(0,x)

렐루 함수는 특정 양수값에 수렴하지 않으므로 깊은 신경망에서 시그모이드 함수보다 훨씬 더 잘 작동합니다. 뿐만 아니라, 어떤 연산이 필요한 것이 아니라 단순 임계값이므로 연산 속도도 빠릅니다. 하지만 입력값이 음수면 기울기도 0이 됩니다. 그리고 이 뉴런은 다시 회생하는 것이 매우 어렵습니다. 이 문제를 죽은 렐루(dying ReLU)라고 합니다.

## 리키 렐루(Leaky ReLU)

f(x) = max(ax,x) (ex. a=0.01과 같이 새는 정도의 기울기) 

죽은 렐루를 보완하기 위해 ReLU의 변형 함수들이 등장하기 시작했습니다.  Leaky ReLU는 입력값이 음수일 경우에 0이 아니라 0.001과 같은 매우 작은 수를 반환하도록 되어있습니다.
