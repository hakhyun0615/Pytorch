# 분산 표현(Distributed Representation)

희소 표현: 고차원에 각 차원이 분리된 표현 방법  
분산 표현: 저차원에 단어의 의미를 여러 차원에다가 분산하여 표현(단어 간 유사도 계산 가능)

# Word2Vec

Word2Vec의 2가지 방식
- CBOW
- Skip-Gram

Word2Vec의 입력: 주변 단어들의 원-핫 벡터  
Word2Vec의 출력: 중심 단어의 원-핫 벡터

# CBOW(Continuous Bag of Words)


- CBOW: 주변 단어들 -> 중간 단어(주변에 있는 단어들을 가지고, 중간에 있는 단어들을 예측하는 방법)  
- 윈도우(window): 중간 단어 양 옆에 볼 단어의 수
- 슬라이딩 윈도우(sliding window): 윈도우를 계속 움직여서 주변 단어와 중심 단어 선택을 바꿔가며 학습을 위한 데이터 셋 만드는 법
- CBOW 학습: Input layer와 Projection layer 사이의 가중치 행렬 W와 Projection layer와 Output layer 사이의 가중치 행렬 W'을 업데이트
- 룩업 테이블(lookup table): one Input layer vector * W(Input layer -> Projection layer 계산과정)
- Projection layer vector: Input layer의 각 원-핫 벡터들이 룩업 테이블을 통해 나온 벡터들의 '평균'  
- Output layer vector: softmax(Projection layer vector * W')
- loss function: (Output layer vector와 중간 단어의 원-핫 벡터 사이의) cross-entropy

## Skip-gram

  
- Skip-gram: 중간 단어 -> 주변 단어들
- 성능이 CBOW보다 좋은 것으로 알려짐
